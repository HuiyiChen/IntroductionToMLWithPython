# Many linear classification models are for binary classification only, and don’t extend naturally to the multiclass case (with the exception of logistic regression).  Common technique to extend a binary classification algorithm to a multiclass classification algorithm is the one-vs.-rest approach. Here a binary model is learned for each class that tries to separate that class from all other classes. So we have as many binary models as there are classes. The classifier that has the highest score on its single class “wins,” and this class label is returned as the prediction.  Formula: w[ 0] * x[ 0] + w[ 1] * x[ 1] + ... + w[ p] * x[ p] + b

```python
import matplotlib.pyplot as plt
import mglearn
from sklearn.datasets import make_blobs

%matplotlib inline
```

```python
X, y = make_blobs(random_state=42)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.legend(["Class 0", " Class 1", "Class 2"])

```
